```python
import pandas as pd
import numpy as np

# data imports
df = pd.read_csv("data/listings/raw")
df.head()
```

```python
# Clean up price column and split in weekly and monthly rent
raw_prices = df.price.values
cleanPrices = []

for raw_price in raw_prices:
    
    cleaned_string = raw_price
    
    # Remove redundent symbols
    cleaned_string = cleaned_string.replace("\n","")
    cleaned_string = cleaned_string.replace("¬£","")
    cleaned_string = cleaned_string.replace(" ","")
    cleaned_string = cleaned_string.replace(",","")
    
    # fix standardise into pcm if pw
    if ("pw" in cleaned_string):
        stringComps = cleaned_string.split("pw")
        if "-" in stringComps[0]: 
            priceRange = stringComps[0].split("-")
            priceRange = [float(price)*4 for price in priceRange]
            low, high = priceRange
        else:
            low, high = [stringComps[0], stringComps[0]]
              
    # sort range 
    elif ("pcm" in cleaned_string):
        stringComps = cleaned_string.split("pcm")
        if "-" in stringComps[0]: 
            priceRange = stringComps[0].split("-")
            low, high = [float(price) for price in priceRange]
        else:
            low, high = [float(stringComps[0]),float(stringComps[0])]

    cleanedPrice = {
        "low": low,
        "high": high
    }
        
    # append
    cleanPrices.append(cleanedPrice)

# Get range into "low", "high" 

print(cleanPrices[0:50])
print("")
print(len(cleanPrices))

```
